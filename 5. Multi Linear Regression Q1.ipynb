# Multi Linear Regression on 50 startup data

Prepare a prediction model for profit of 50_startups data. Do transformations for getting better predictions of profit and
make a table containing R^2 value for each prepared model.

R&D Spend -- Research and devolop spend in the past few years

Administration -- spend on administration in the past few years

Marketing Spend -- spend on Marketing in the past few years

State -- states from which data is collected

Profit  -- profit of each state in the past few years

# Preparing a model

### Using Statsmodels

```python
#Build model
import statsmodels.formula.api as smf 
model1 = smf.ols('profit~rd_spend+admin+mkt_spend+state',data=data).fit()
```

```python
model1.summary() 
```

The model summary above reveals that only rd_spend truly is a factor in profit. The remaining variable have p-value greater than 0.05 which means we have to accept the null hypothesis that the factors do not contribute to profit. 

We can a try a model with rd_spend and other variablessee the results.

# Simple Linear Regression models 

We can try a number of SLR models to determine the f-statistics for each variable.


 <h4>rd_spend and profit</h4>

```python
model2 = smf.ols('profit~rd_spend',data=data).fit()
model2.summary() 
```

<h5>rd_spend alone contributes 94.5% variance to profit which is very high and enough to determine the profit. </h5>

 <h4>mkt_spend and profit</h4>

```python
model3 = smf.ols('profit~mkt_spend',data=data).fit()
model3.summary() 
```

mkt_spend alone contributes 55% variance to profit which is good. 


<h4>Model using mkt_spend and rd_spend</h4>

```python
model4 = smf.ols('profit~rd_spend+mkt_spend',data=data).fit()
model4.summary()
```

Together, they got 95% of the variance, which is a promising outcome, albeit with some minor autocorrelation between the two. In the subsequent steps, I plan to conduct residual analysis and Model Deletion Diagnostics involving both Model 4 and Model 1 to discern which one exhibits superior performance."

```python
rsq_rdspend = smf.ols('rd_spend~mkt_spend+admin',data=data).fit().rsquared  
vif_rdspend = 1/(1-rsq_rdspend)

rsq_mktspend = smf.ols('mkt_spend~rd_spend+admin',data=data).fit().rsquared  
vif_mktspend = 1/(1-rsq_mktspend) 

rsq_admin = smf.ols('admin~mkt_spend+rd_spend',data=data).fit().rsquared  
vif_admin = 1/(1-rsq_admin) 

# Storing vif values in a data frame
d1 = {'Variables':['rd_spend','mkt_spend','admin'],'VIF':[vif_rdspend,vif_mktspend,vif_admin]}
Vif_frame = pd.DataFrame(d1)  
Vif_frame
```

#### Residual Plot for Homoscedasticity 

```python
def get_standardized_values( vals ):
    return (vals - vals.mean())/vals.std()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Residual plot for model1
ax1.scatter(get_standardized_values(model1.fittedvalues), get_standardized_values(model1.resid))
ax1.set_title('Residual Plot for Model 1')
ax1.set_xlabel('Standardized Fitted values')
ax1.set_ylabel('Standardized residual values')

# Residual plot for model4
ax2.scatter(get_standardized_values(model4.fittedvalues), get_standardized_values(model4.resid))
ax2.set_title('Residual Plot for Model 4')
ax2.set_xlabel('Standardized Fitted values')
ax2.set_ylabel('Standardized residual values')

# Adjust subplot layout
plt.tight_layout()

# Display the plots
plt.show()
```

model4 distribution is very similar to model1

## Model Deletion Diagnostics

#### Detecting Influences

#### Cook's Distance -Model1

```python
model_influence = model1.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
#Plot the influencers values using stem plot
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(data)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show() 
```

### Cook's Distance -Model4

```python
model_influence = model4.get_influence()
(c, _) = model_influence.cooks_distance
```

```python
fig = plt.subplots(figsize=(20, 7))
plt.stem(np.arange(len(data)), np.round(c, 3))
plt.xlabel('Row index')
plt.ylabel('Cooks Distance')
plt.show() 
```

```python
#index and value of influencer where c is more than 0.35
(np.argmax(c),np.max(c))
```

### Dropping outliers 

```python
data_new = pd.read_csv("50_Startups.csv")
data1=data_new.drop(data_new.index[[46,48,49]],axis=0).reset_index()
data1=data1.drop(['index','Administration','State'],axis=1)
data1.rename(columns={'R&D Spend':'rd_spend','Marketing Spend':'mkt_spend','Profit':'profit'  }, inplace = True)
data1.head()
```

# Building final Model

We will make the final model with cleaned data and model4 variables.

```python
fmodel = smf.ols('profit~rd_spend+mkt_spend',data=data1).fit()
fmodel.summary() 
```

Best values of R-Square. AIC values have come down.so  this will be the best model.
